```python
### Calculate forward links ###
# Calculate Leontief Inverses for all cells
# Calculate row sums of Leontief Inverses
# Result is the forward linkage of each sector
```

```python
import pandas as pd
import numpy as np
import os
```

```python
# Load with correct indexation
A_df_gas = pd.read_csv("C:/Users/danie/Nextcloud/Coding/Masterthesis/notebooks/NB_exio3_figaro_gas/figaro_B_modified.csv", index_col=[0, 1], header=[0, 1])
```

```python
display(A_df_gas)
```

```python
# Calculate forward linkages using Leontief inverse

# Step 1: Create identity matrix matching dimensions of df
identity_matrix = np.eye(A_df_gas.shape[0])

# Step 2: Calculate Leontief inverse
A_matrix = A_df_gas.values  # Technical coefficients as numpy array
leontief_inverse = np.linalg.inv(identity_matrix - A_matrix)

# Reconstruct a DataFrame to preserve indices
L_df = pd.DataFrame(leontief_inverse, index=A_df_gas.index, columns=A_df_gas.columns)

# Step 3: Calculate forward linkages by summing across rows
forward_linkages = L_df.sum(axis=1)

# Convert to DataFrame for clearer representation
forward_linkages_df = forward_linkages.to_frame(name='Forward_Linkage')

# Display the forward linkages neatly
display(forward_linkages_df)

# Save the forward linkages to a CSV file
forward_linkages_df.to_csv("C:/Users/danie/Nextcloud/Coding/Masterthesis/notebooks/NB_exio3_figaro_gas/forward_linkages.csv")
```
