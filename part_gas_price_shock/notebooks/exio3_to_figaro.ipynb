```python
### Conversion of FIGARO tables to new table with gas and other energy seperated ###

import pandas as pd

# 1️⃣ 📌 Load the FIGARO dataset first
figaro_file_path = r"C:\Users\danie\Nextcloud\Coding\Masterthesis\data\raw\figaro_tables\2021_figaro_64.csv"
figaro_2021 = pd.read_csv(figaro_file_path, index_col=0)  # Set row labels as index

# 2️⃣ 📌 Extract the Country Code and Sector Name Separately
def split_country_sector(label):
    """Splits country-sector into (country, sector) tuple."""
    parts = label.split("_", 1)  # Splitting at the first underscore
    return parts[0], parts[1] if len(parts) > 1 else parts[0]  # If no sector, return as-is

# Split row and column labels into country & sector
row_info = figaro_2021.index.map(split_country_sector)
col_info = figaro_2021.columns.map(split_country_sector)

# Convert to DataFrame for easy mapping
row_df = pd.DataFrame(row_info.tolist(), index=figaro_2021.index, columns=["Country", "Sector"])
col_df = pd.DataFrame(col_info.tolist(), index=figaro_2021.columns, columns=["Country", "Sector"])

# 3️⃣ 📌 Define the sector mapping AFTER loading and extracting the country-sector
figaro_mapping = {
    "A01": "A01",
    "A02": "A02",
    "A03": "A03",
    "B": ["B_gas", "B_nongas"],  # Split into Gas and Non-Gas
    "C10T12": "C10-12",
    "C13T15": "C13-15",
    "C16": "C16",
    "C17": "C17",
    "C18": "C18",
    "C19": "C19",
    "C20": "C20-21",
    "C21": "C20-21",
    "C22": "C22",
    "C23": "C23",
    "C24": "C24",
    "C25": "C25_33",  # Grouped
    "C26": "C26",
    "C27": "C27",
    "C28": "C28_32",  # Grouped with C32
    "C29": "C29",
    "C30": "C30",
    "C31_32": "C31",  # C32 is included in C28_32
    "C33": "C25_33",  # Grouped with C25
    "D35": "D35",
    "E36": "E36",
    "E37T39": "E37-39",
    "F": "F",
    "G45": "G45",
    "G46": "G46",
    "G47": "G47",
    "H49": "H49",
    "H50": "H50",
    "H51": "H51",
    "H52": "H52",
    "H53": "H53",
    "I": "I",
    "J58": "J62_63",
    "J59_60": "J62_63",
    "J61": "J62_63",
    "J62_63": "J62_63",
    "K64": "K64",
    "K65": "K65",
    "K66": "K66",
    "L": "L68",  # Real Estate
    "M69_70": "M_N",
    "M71": "M_N",
    "M72": "M_N",
    "M73": "M_N",
    "M74_75": "M_N",
    "N77": "M_N",
    "N78": "M_N",
    "N79": "M_N",
    "N80T82": "M_N",
    "O84": "O84",
    "P85": "P85",
    "Q86": "M_N",
    "Q87_88": "M_N",
    "R90T92": "R_S",
    "R93": "R_S",
    "S94": "R_S",
    "S95": "R_S",
    "S96": "R_S",
    "T": "T",
    "U": "U"
}

# 4️⃣ 📌 Apply Mapping to Modify Only the Sector Part
row_df["Sector"] = row_df["Sector"].replace(figaro_mapping)
col_df["Sector"] = col_df["Sector"].replace(figaro_mapping)

# Reconstruct new country-sector labels
row_df["New_Label"] = row_df["Country"] + "_" + row_df["Sector"]
col_df["New_Label"] = col_df["Country"] + "_" + col_df["Sector"]

# Apply new labels to the dataset
figaro_2021.index = row_df["New_Label"]
figaro_2021.columns = col_df["New_Label"]

# 5️⃣ 📌 Aggregate Data by Summing Over Mapped Sectors
figaro_aggregated = figaro_2021.groupby(figaro_2021.index).sum()
figaro_aggregated = figaro_aggregated.T.groupby(figaro_aggregated.T.columns, axis=1).sum()

# ✅ **Ensure Symmetry**
assert figaro_aggregated.shape[0] == figaro_aggregated.shape[1], "Row-column mismatch after aggregation!"

# ✅ **Check Energy Sector Per Country**
print("Aggregated Energy Sectors Per Country:")
energy_countries = [label for label in figaro_aggregated.index if "B_" in label]
print(figaro_aggregated.loc[energy_countries, energy_countries])

# 📌 **Save the processed file**
figaro_aggregated.to_csv("figaro_aggregated_2021.csv")
```
